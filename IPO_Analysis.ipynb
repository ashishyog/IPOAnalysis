```python
# imports
import pandas as pd
import numpy as np
import os
```

```python
# read the data with all default parameters
data = pd.read_csv('C:\\Users\\ashishyog\\Desktop\\ML_Project\\data\\IPO-Data_1.csv', encoding='Latin-1')
```

```python
data
```

```python
def performanceValue(change):
        if (change>10):
            return 'excellent'
        elif (change<=10 and change>0):
            return 'good'
        else :
            return 'bad'
        
```

```python
# replace the missing values
data["Performed"]=data["PxChng"].apply(performanceValue)
```

```python
data
```

```python
data
```

```python
# one hot encoding for Managers
data = pd.get_dummies(data,columns=['Manager1'])
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size = 0.3)
```

```python
train.head()
```

```python
test.head()
```

```python
data.columns = [c.replace('Manager1_', '') for c in data.columns]
data.columns = [c.replace(' ', '_') for c in data.columns]
```

```python
trainColumns = list(filter(lambda x: x != "Performed", list(data)))
trainColumns
```

```python
data = data.reindex(columns=trainColumns)
data
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(max_leaf_nodes=25)
clf=clf.fit(train[["IPOPrice", "OpenPrice", "ClosePrice", "PxChng", "OpeningChng", "Ratings"]],train["Performed"])
```

```python
clf
```

```python
clf.feature_importances_
```

```python
predictions = clf.predict(test[["IPOPrice", "OpenPrice", "ClosePrice", "PxChng", "OpeningChng", "Ratings"]])
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(test["Performed"], predictions)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=1000,max_leaf_nodes=15)
```

```python
def checkAccuracy(clf):
    clf=clf.fit(train[["IPOPrice", "OpenPrice", "ClosePrice", "PxChng", "OpeningChng", "Ratings"]],train["Performed"])
    predictions = clf.predict(test[["IPOPrice", "OpenPrice", "ClosePrice", "PxChng", "OpeningChng", "Ratings"]])
    return accuracy_score(test["Performed"], predictions)
```
